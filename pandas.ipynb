# merging

```python
import pandas as pd

df1 = pd.DataFrame({'HPI':[80,90,70,60], 'IND_GDP':[50,45,45,67],'Int_rate':[2,1,2,3]}, index=[2001,2002,2003,2004])
df1
df2 = pd.DataFrame({'HPI':[80,90,70,60], 'IND_GDP':[50,45,45,67], 'Int_rate':[2,1,2,3]}, index=[2005,2006,2007,2008])
df2

merge = pd.merge(df1, df2) # here some columns will be common
print(merge)

merge = pd.merge(df1, df2, on='HPI')
print(merge)
```

# joining


```python
df1 = pd.DataFrame({'HPI':[80,90,70,60], 'IND_GDP':[50,45,45,67], 'Int_rate':[2,1,2,3]}, index=[2001,2002,2003,2004])

df2 = pd.DataFrame({'Low_Tier_HPI':[50,45,67,34], 'unemployment':[1,3,5,6]}, index = [2001,2003,2004,2004])

joined = df1.join(df2)
print(joined)
```

# changing the column names


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")
df = pd.DataFrame({'Day':[1,2,3,4], 'Visitors':[200,100,230,300], 'Bounce_Rate':[20,45,60,10]})
df.set_index("Day", inplace=True) # the index is changed to day


# changing the column header or simply renaming it
df = df.rename(columns={'Visitors':'Users'})
print(df)

df.plot()
plt.show()
```

# concatenation

```python
import pandas as pd

df1 = pd.DataFrame({'HPI':[80,85,88,85], 'Ink_rate':[2,3,2,2], 'US_GDP_Thousands':[50,55,65,55]}, index=[2001,2002,2003,2004])
df2 = pd.DataFrame({'HPI':[80,85,88,85], 'Ink_rate':[2,3,2,2],'US_GDP_Thousands':[50,55,65,55]}, index=[2005,2006,2007,2008])
print(df1)
print(df2)

concat = pd.concat([df1,df2])
print(concat)
```
