# fourier transformation
-this is a method that deals with expressing a function as a sum of period components and recovering the signals from those 3components
the fft and ifft can be used can be used to return the discrete fourier of a real or complex sequence


```python
from scipy.fftpack import fft,ifft
import numpy as np
x = np.array([(1,2,3,4)])
y = fft(x)
print(y)
```

```python
# inverse
y = ifft(x)
print(y)
```

# interpolation functions
-this refers to contructing new data points within a set of know data points. The scipy interpolate consist of spline functions and classes, one dimensional and multi-dimensional(univariate and multivaliate) interpolation classes 

```python
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.arange(5, 20)
y = np.exp(x/3.0)
f = interpolate.interp1d(x, y)
x1 = np.arange(6, 12)
y1 = f(x1) # use interpolation function returned by interp1d
plt.plot(x, y, 'o', x1, y1, '--')
plt.show()
```
