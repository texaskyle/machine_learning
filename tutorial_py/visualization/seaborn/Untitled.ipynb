```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style='darkgrid')   # setting the background 
```

```python
# loading the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# drawing a relation plot
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
tips['smoker'].value_counts()
```

```python
tips['total_bill'].value_counts()
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='time', data=tips)
```

```python
# using the size column 
sns.relplot(x='total_bill', y='tip', hue='time', style='size', data=tips)
```

```python
# size
sns.relplot(x='total_bill', y='tip', hue='smoker', style='time', size='size' ,data=tips)
```

## creating a new dataframe

```python
import numpy as np
from numpy.random import randn
```

```python
df7 = pd.DataFrame(dict(time=np.arange(500), value=randn(500).cumsum()))
df7.head(5)
```

```python
sns.relplot(x='time', y='value', kind='line', data=df7, sort=True)
```

```python
## family dataset
fmri = sns.load_dataset('fmri')
fmri.head(5)
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, ci='sd')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event' ,style='region', markers=True)
```
