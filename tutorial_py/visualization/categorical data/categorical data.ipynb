## joint plot

```python
# using the tips data set
tips = sns.load_dataset('tips')
tips
```

```python
x = tips['total_bill']
y = tips['tip']
```

```python
sns.jointplot(x=x, y=y)
```

```python
sns.jointplot(x=x, y=y, kind='hex')
```

```python
sns.jointplot(x=x, y=y, kind='kde')
```

```python
import matplotlib.pyplot as plt

g = sns.jointplot(x=x, y=y, kind='kde', color='r')
g.plot_joint(plt.scatter, c='r', s=10, linewidth=7, marker='+')
g.ax_joint.collections[0].set_alpha(0)
```

## sinplot()

```python
import numpy as np
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i*0.5) * (7-i))
```

```python
sinplot()
```

```python
def sinshowplot():
    x = np.linspace(0, 14, 100)
    for i in range(3):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i))
```

```python
sinshowplot()
```
