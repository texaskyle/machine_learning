```python
# loading a dataset
pd.read_csv('IRIS.csv')
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris.plot(figsize=(10,5), title='iris-dataset')
plt.ylabel('length')
plt.xlabel('X Axis')
```

```python
ts
```

```python
ts.plot(kind='bar')
```

```python
# beacause species is categorical data i may not be able to plot it, so it need to be dropped
df = iris.drop('species', axis=1)
df
```

```python
# picking one row of this dataset
df.iloc[0]
```

```python
# ploting this row
df.iloc[0].plot(kind='bar', figsize=(7,5))
plt.xlabel('column names')
plt.ylabel('magnitude')
plt.legend()
plt.grid()
plt.title('row one plot')
```
