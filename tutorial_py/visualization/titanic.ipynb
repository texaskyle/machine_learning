# scatter matrix
a scatter matrix tries to show the relationship of every dataset with every dataset

the diagnaol matrix represent the variance of each variable

the off diagnal represent the relationship between diffrent variables

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(iris, figsize=(6,6), diagonal='kde', color='b')
plt.show()
```

```python
# if you want to know if a graph is normalized of a certain column 
iris['petal_length'].plot.kde()
```

```python
# checking whether the data of a certain column is normalized or not
iris['petal_width'].plot.kde()
```
