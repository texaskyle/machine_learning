```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# reading the dataset
suv_data = pd.read_csv('C:\\Users\\charles\\Desktop\\Jupiter\\edureka\\LogisticRegresion\\suv_predictions\\suv_data.csv')
print(suv_data.head(10))

# how many users in the dataset
print(f"users = {len(suv_data.index)}")
```

# Analyzing the data

```python
# checking the people who bought the car
sns.countplot(x='Purchased', data=suv_data)
plt.grid()
plt.title("purchases")
```

```python
# analyzing how many males and females who purchased the cars
sns.countplot(x='Purchased', hue='Gender', data=suv_data)
```

```python
# using a histogram to plot the age of the users
suv_data['Age'].plot.hist(bins=20, figsize=(10,5))
plt.xlabel('Age')

# maximum age
max_age = max(suv_data['Age'])
print(f"maximum age: {max_age}")
```

```python
# histogram plot of salary distribution
suv_data['EstimatedSalary'].plot.hist()
plt.title("Salary distribution")
plt.xlabel('salary')
```

```python
# getting all the info, to check the rows which are empty
suv_data.info()
```

## Data wrangling
- this is cleaning the data

```python
suv_data.isnull()
```

```python
# counting the sum of the data which can be null
suv_data.isnull().sum()
```

```python
# to get a visualization of the data which can be null
sns.heatmap(suv_data.isnull(), yticklabels=0, cmap='viridis')
```

```python
# getting the dummies of gender
sex = pd.get_dummies(suv_data['Gender'], drop_first=True)
```

```python
sex.head(5)
```

```python
# concatenating with the new column sex
suv_data = pd.concat([suv_data, sex], axis=1)
```

```python
suv_data.head(5)
```

```python
# dropping all the irrelevant column
suv_data.drop('User ID', axis=1, inplace=True)
```
