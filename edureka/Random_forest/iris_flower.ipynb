```python
import pandas as pd
import numpy as np
# setting the random seed
np.random.seed(0)
# laoding the iris dataset
from sklearn.datasets import load_iris
# importing the random forest module
from sklearn.ensemble import RandomForestClassifier
```

```python
# create an object called iris with iris data
iris = load_iris()
print("iris")

# creating a dataframe with four feature variables
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(10)
```

```python
# adding another column for the species names
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df[70:100]
```

```python
print(iris)
```

```python
# creating the testing and the training data
df['is_train'] = np.random.uniform(0, 1, len(df)) <= 0.75
df.head(6)
```

```python
# creating the testing and the training variables
train, test = df[df["is_train"]==True] , df[df['is_train']==False]

# print the training data
print("The training data lenght is:: ",len(train))
print("The training data lenght is:: ",len(test))
```

```python
# creating a list of column names
features = df.columns[:4]
```

```python
# converting each species name into digits
y = pd.factorize(train['species'])[0]
y
```

## Creating the model

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_jobs=2, random_state=2)
```

```python
# fitting the model the data
clf.fit(train[features], y)
```

```python
# testing the trained model
clf.predict(test[features])
```

```python
# viewing the predicted probability for the first 10-20 observations
clf.predict_proba(test[features])[10:20]
```

```python
# mapping the names for the plants for each predicted plant class
pred = iris.target_names[clf.predict(test[features])]
pred[:5]
```

```python
# viewing the actual species for the first 5 observation
test['species'].head(5)
```

```python
# creating a confusion matrix
pd.crosstab(test['species'], pred, rownames=['Actual_species'], colnames=['predicted species'])
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test['species'], pred)
```

```python
# doing the actual prediction on the data that we dont know which flowers 
preds = iris.target_names[clf.predict( [[5.1, 3.5, 1.4, 0.2], [70, 5.9, 3.2, 4.8]] )]

print(preds)

"""sepal length (cm)	sepal width (cm)	petal length (cm)	petal width (cm)
0	5.1	3.5	1.4	0.2
1	4.9	3.0	1.4	0.2
2	4.7	3.2	1.3	0.2
3	4.6	3.1	1.5	0.2
4	5.0	3.6	1.4	0.2"""
```
