```python
import numpy as np
speed = [86, 87, 88, 86, 87, 85, 86]

#calculating the stardard deviation
stardard_deviation = np.std(speed)
print("stardard_deviation: " + str(stardard_deviation))

#calculating the variance
variance = np.var(speed)
print('variance: ' + str(variance))
```

```python
# drawing a histogram
import matplotlib.pyplot as plt

# generating random values
random_floating_numbers = np.random.uniform(0.0, 5.0, 250)
plt.hist(random_floating_numbers, 5)
plt.show()
```
