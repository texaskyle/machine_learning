```python
import matplotlib.pyplot as plt
import numpy 

x = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100]

# method to make polynomial model
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

from sklearn.metrics import r2_score
print(f"relationship: {r2_score(y, mymodel(x))}")

#specifing how the line will be displayed from 1 to 22 in the x-axis and a 100 in the y-axis
myline = numpy.linspace(1, 22, 100)

# drawing the original scatter plot
plt.scatter(x, y)

#drawing the line of polynomial regression
plt.plot(myline, mymodel(myline))

#display the diagram
plt.show()
```
