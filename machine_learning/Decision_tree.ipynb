```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# reading the csv file
df = pd.read_csv('Comedy_data.csv')

# To make a decison tree all data has to be numerical
dictionary = {'UK':0,
    'USA':1,
    'N':2}
df['Nationality'] = df["Nationality"].map(dictionary)

dictionary = {'YES':1,
              'NO':0}
df['Go'] = df['Go'].map(dictionary)

# separating the independent variables and the dependent variables
features = ['Age', 'Experience', 'Rank', 'Nationality']
X = df[features]
y = df['Go']


# creating the actual decison tree
dtree = DecisionTreeClassifier()
dtree =dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)

prediction = dtree.predict([[50, 10, 10, 0]])
print(prediction)
if prediction == 0:
    print('he will not go to the comedy')
else:
    print('he will go to the comedy')
```
