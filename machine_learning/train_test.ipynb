```python
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
numpy.random.seed(2) #seed is used to make sure that the randomly produced values dont keep on changing so that there are reproducible

# dataset
x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100)/x

# spliting the dataset into the training and testing dataset
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

# creating a polynomial model 
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))
myline = numpy.linspace(0, 6, 100)

#determining the rlshp of train_x and train_y
r2_train = r2_score(train_y, mymodel(train_x))
print(f'ralationship of train_y and train_x: {r2_train}')

# determining the relationship btwn test_x and test_y
r2_test = r2_score(test_y, mymodel(test_x))
print(f'the relationship btwn test_y and test_x{r2_test}')

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()

# predicting how much will the coustomer spend if she stays in the shop for 5 minutes
print(mymodel(4))
```
