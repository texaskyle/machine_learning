```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage


x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

# computing the linkage between all the diffrent points. here we use simple euclidean distance to measure and ward's linkage, which seeks to minimize the distance between clusters 
linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data) # ploting the dendrogram. this will show the hierachy of clusters from the botton to the top

plt.scatter(x,y)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

# creates one list, and the data is turned into set points
data = list(zip(x, y))

# AgglomerativeClustering is initialized with 2 clusters, using the same distance linkage and ward linkage
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
# .fit_predict method can be called on the data to compute the clusters using diffrent parameters across our choosen number of parameters
labels = hierarchical_cluster.fit_predict(data)
print(labels)

plt.scatter(x, y, c=labels)
plt.show()
```
